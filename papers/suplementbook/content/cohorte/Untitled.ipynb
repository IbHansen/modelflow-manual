```python
import pandas as pd
import numpy as np
```

```python
# Define dimensions
sexes = ['female', 'male']
ages = [f'age_{i}' for i in range(101)]  # age_0 ... age_100
year = 2024

# Build the data dict
data = {}

# Population by age and sex (stylized declining curve)
for sex in sexes:
    for i, age in enumerate(ages):
        base_pop = max(100000 * np.exp(-i / 25), 1000)
        if sex == 'male':
            base_pop *= 1.05 if i < 50 else 0.95
        data[f'pop__{sex}__{age}'] = [base_pop]

# Mortality rates (increase steeply with age)
for sex in sexes:
    for i, age in enumerate(ages):
        mort_rate = 0.001 + (i / 100) ** 5
        mort_rate = min(mort_rate, 1.0)
        data[f'death_rate__{sex}__{age}'] = [mort_rate]

# Fertility rates (only females aged 15–49)
for sex in sexes:
    for i, age in enumerate(ages):
        if sex == 'female' and 15 <= i <= 49:
            fert_rate = np.exp(-((i - 30) ** 2) / (2 * 7 ** 2)) * 0.1
        else:
            fert_rate = 0.0
        data[f'fertility__{sex}__{age}'] = [fert_rate]

# Create final DataFrame
df = pd.DataFrame(data, index=[year])
df.T
```
